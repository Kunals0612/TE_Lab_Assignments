```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./Datasets/titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
df['Cabin'].fillna("unknown", inplace = True)
df['Embarked'].fillna("unknown", inplace = True)
```

```python
df.isnull().sum()
```

```python
unique_values = df['Pclass'].unique()
print(unique_values)
```
