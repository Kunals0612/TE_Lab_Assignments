```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Datasets/titanic.csv')
df.head()
```

```python
survived_df = df[df['Survived'] == 1]
sb.countplot(x=df['Sex'], data = survived_df)
plt.title('Total Number of Survivors by Gender')
plt.xlabel('Gender')         
plt.ylabel('Number of Survivors')  
```

```python
sb.countplot(x=df['Pclass'], data = survived_df)
plt.title('Total Number of Survivors by Passanger Class')
plt.xlabel('PClass')         
plt.ylabel('Number of Survivors') 
```

```python
bins = [0, 12, 18, 35, 50, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior', 'Elderly']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
sb.countplot(x=df['AgeGroup'], data=survived_df)
```

```python
sb.countplot(x=df['Embarked'], data = survived_df)
plt.title('Survival by Emabarked Location')
plt.xlabel('Embarkation Location')
plt.ylabel('Survival Count')
```

```python
sb.violinplot(x=df['Pclass'], y=df['Age'])
```

```python
sb.swarmplot(x = df['Age'], data = survived_df)
```

```python
sb.histplot(data = df['Fare'], kde = True)
```
