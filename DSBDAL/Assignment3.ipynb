```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./Datasets/Iris.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
dfg = df.groupby('Species')
dfg.describe()
```

```python
columns = ["SepalLengthCm","SepalWidthCm",	"PetalLengthCm",	"PetalWidthCm"]
print(columns)
mean = []
median = []
var = []

dfgrouped = df.groupby('Species')

for species, group in dfgrouped:
    print(f"statistics for: {species}")
    for col in columns:
        val = group[col]
        variance = sum((val - val.mean())**2) / len(val)
        deviation = variance**0.5
        print(f"Column: {col}")
        print(f"Variance = {variance}")
        print(f"Standard Deviation = {deviation}")
        print(f"Mean = {val.mean()}")
        print(f"Median = {val.median()}")
        print(f"Inter Quartile Range = {val.quantile(0.75) - val.quantile(0.25)}")
        print(f"Min = {val.min()}")
        print(f"Max = {val.max()}")
        print("\n")
        mean.append(val.mean())
        median.append(val.median())
        var.append(variance)
```

NBA Dataset

```python
df = pd.read_csv("./Datasets/nba.csv")
df.head()
```

```python
df["College"]=df["College"].fillna("No College")
df["Salary"]=df["Salary"].fillna(df["Salary"].mean())
df=df.dropna()
df.isna().sum()
```

```python
df.describe()
```

```python
df['Age'] = df['Age'].astype('object')
dfg = df.groupby('Age')
```

```python
columns = ["Salary"]
print(columns)

for ageGroup, group in dfg:
    print(f"Statistics for {ageGroup}:\n")
    
    for col in columns:
        val = group[col]  # Get values for the current column of the species
        
        # Calculate variance, standard deviation, mean, median, etc.
        variance = sum((val - val.mean())**2) / len(val)
        deviation = variance**0.5
        
        print(f"Column: {col}")
        print(f"Variance = {variance}")
        print(f"Standard Deviation = {deviation}")
        print(f"Mean = {val.mean()}")
        print(f"Median = {val.median()}")
        print(f"Inter Quartile Range = {val.quantile(0.75) - val.quantile(0.25)}")
        print(f"Min = {val.min()}")
        print(f"Max = {val.max()}")
        print("\n")
```
